# list methods

```python
Employee = ["Ayush" , "Sajeed" , "Khushi" , "Siddhi" , "Saptesh" , "vinayak"]

Employee[0] = "Nishant"

print(Employee)
```

```python
Employee.insert(7,"Sajeed")
print(Employee)
```

```python
Employee.remove("Sajeed")
print(Employee)
```

```python
Employee.pop(0)
print(Employee)
```

```python
a  = Employee.index("vinayak")
print(a)
```

```python
c = Employee.count("vinayak")
print(c)
```

```python
Employee.sort()
print(Employee)
```

```python
for items in enumerate(Employee):
    print(items)
```

```python

z = len(Employee)
print(z)
```

# List comprehension

```python
number = []
for i in range(11):
    number.append(i)

print(number)
```

```python
number = [x for x in range(1,51)]
```

```python
print(number)
```

```python
list_pract = [x**2 for x in range(1,19) if x%2==0]
```

```python
print(list_pract)
```
